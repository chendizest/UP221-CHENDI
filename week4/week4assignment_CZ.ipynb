# Data Management

```python
# import all the tool
import pandas as pd
import geopandas as gpd
```

```python
# read in a geojson file 
tracts = gpd.read_file('California_County_Boundaries.geojson')
tracts.head()
```

```python
tracts
```

```python
# load wildfire planning effective score csv
df = pd.read_csv('PE_58CAcounties.csv')
df
```

```python
# manipulate wildfire planning effective score csv as needed
columns_to_drop = df.columns[59:73] 
columns_to_drop
df.drop(columns_to_drop, axis=1, inplace=True)
df
```

```python
TRANSPOSED_df = df.T
TRANSPOSED_df.head()
```

```python
TRANSPOSED_df.columns = TRANSPOSED_df.iloc[0]
TRANSPOSED_df2 = TRANSPOSED_df.drop(TRANSPOSED_df.index[0])
TRANSPOSED_df2
```

```python
TRANSPOSED_df2['County_FIPS_ID'] = tracts['County_FIPS_ID'].values
TRANSPOSED_df2
```

```python
TRANSPOSED_df2['3 PLANS TOTAL-PROPORTIONATE'] = pd.to_numeric(TRANSPOSED_df2['3 PLANS TOTAL-PROPORTIONATE'], errors='coerce')
```

```python
TRANSPOSED_df2['3 PLANS TOTAL-PROPORTIONATE'].describe()
```

```python
TRANSPOSED_df2['3 PLANS TOTAL-PROPORTIONATE'].plot.hist(bins=50)
```

```python
TRANSPOSED_df2.boxplot(column=['3 PLANS TOTAL-PROPORTIONATE'])
```

```python
df_sorted = TRANSPOSED_df2.sort_values(by='3 PLANS TOTAL-PROPORTIONATE',ascending = False)
df_sorted.head(10).plot.bar(x='County_FIPS_ID',
                            y='3 PLANS TOTAL-PROPORTIONATE'
                          )
```

# Mapping

```python
# plot it
tracts.plot(figsize=(12,10))
```

```python
mapping = tracts.merge(TRANSPOSED_df2, on="County_FIPS_ID")
mapping.head()
```
